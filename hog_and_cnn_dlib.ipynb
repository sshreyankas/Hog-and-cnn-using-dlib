```python
import cv2
import dlib
import argparse
import time
```

```python
from imutils import face_utils
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/home/shreyanka/Documents/Clique/friends.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
face = dlib.get_frontal_face_detector()
rects = face(gray, 1)

for(i,rects) in enumerate(rects):
  (x,y,w,h) = face_utils.rect_to_bb(rects)

  cv2.rectangle(img, (x,y), (x+w, y+h), (255,0,0), 2)

plt.imshow(img)
plt.show()
```

```python
img = cv2.imread('/home/shreyanka/Documents/Clique/Input.jpg',0)
dnnFaceDetector = dlib.cnn_face_detection_model_v1("/home/shreyanka/Documents/Clique/mmod_human_face_detector.dat")
faces_cnn = dnnFaceDetector(img, 1)
for faceRect in faces_cnn:
    x1 = faceRect.rect.left()
    y1 = faceRect.rect.top()
    x2 = faceRect.rect.right()
    y2 = faceRect.rect.bottom()
    cv2.rectangle(img, (x1,y1), (x2,y2), (0,0,255), 2)
img_height, img_width = img.shape[:2]
cv2.putText(img, "CNN", (img_width-50,40), cv2.FONT_HERSHEY_SIMPLEX, 0.5,(0,0,255), 2)
# cv2.imshow("cnn face detection with dlib", img)
plt.imshow(img)
plt.show()
# cv2.waitKey()
# cv2.destroyAllWindows()
```
